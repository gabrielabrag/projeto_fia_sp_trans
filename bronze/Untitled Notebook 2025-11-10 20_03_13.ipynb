```python
import plotly.express as px

# Consulta os dados do Databricks SQL
df = spark.sql("""
SELECT 
  veiculo_prefixo, 
  lat, 
  lon, 
  timestamp_api
FROM sp_trans2.gold.fato_posicao
WHERE veiculo_prefixo = '55378'
ORDER BY timestamp_api
""")

# Converte para Pandas
df_pd = df.toPandas()

# Cria o mapa com a linha conectando os pontos
fig = px.line_mapbox(
    df_pd,
    lat="lat",
    lon="lon",
    hover_name="timestamp_api",
    color_discrete_sequence=["blue"],
    zoom=12,
    height=600
)

# Configura estilo do mapa
fig.update_layout(mapbox_style="open-street-map", title="Rota do ve√≠culo 55378")

fig.show()
```
